# MLB is one of the hardest sports leagues to make the Playoffs!
## Out of the 30 MLB Team (15 in Teams in per League)
## Only 8 out of 30 Teams make the Playoffs (27%)
## 6 out of the 8 Teams make the Playoffs by virtue of winning their Division (75%)
## The remaining 2 Teams make the playoffs through the Wild Card, 
###  - which is the Team with the best record of all remaining Teams
## Over the past 10 Seasons,  16 out of the 20 Teams to make the World Series won their respective Division (80%)
## And over the past 10 Season, 7 out of the past 10 World Series Winners won their respective Division (70%)

# Goal for Today is to indentify the following:
## 1) Are there any Divisions that are easier to win(to make the Playoffs)? 
## 2) Are there any Divisions that are more competitive than other (Games Back)?
## 3) Is there a 'Magic Number of Wins' to finished 1st in a specific Division?

```python
#https://medium.com/better-programming/using-the-mlb-stats-api-to-get-daily-data-88f48266656c

def get_division_standings(season, league, division):
    standings = pd.DataFrame(statsapi.standings_data(league, season=season)[division]['teams'])
    standings = standings[['name', 'div_rank', 'gb', 'w', 'l']]
    standings['Season'] = season
    standings['Division'] = division
    #standings.to_csv(header=False,index=False)
    return standings
get_division_standings(2014,103, 201)
```

```python
all_standings=pd.DataFrame()
seasons = (2010, 2011,2012,2013,2014,2015,2016,2017,2018,2019)

leagues ={103: [200, 201, 202],
           104: [203, 204, 205]
         }
for season in seasons:
    
    for league in leagues:
        for division in leagues[league]:
                standings = pd.DataFrame(statsapi.standings_data(league, season=season)[division]['teams'])
                col_list = ['name', 'div_rank', 'gb', 'w', 'l']
                standings = standings[col_list]
                standings['Season'] = season
                standings['Division Code'] = division
                all_standings = all_standings.append(standings)
```

```python
all_standings.head(9)
```

```python
all_standings['Division'] = all_standings['Division Code']
```

```python
all_standings.info()
```

```python
all_standings.rename(
    columns = {
    'name':'Team Name',
    'div_rank':'Division Rank',
    'gb':'Games Back', 
    'w': 'Wins',   
    'l': 'Loses'           
    },
    inplace =True
)
```

```python
all_standings.head(5)
```

```python
all_standings = all_standings[["Season", "Division Code", "Division", "Division Rank", "Team Name", "Wins", "Loses", "Games Back"]]
```

```python
all_standings.head(5)
```

```python
all_standings.info()
```

```python
#200 = AL West 
#201 = AL East
#202 = AL Central
#203 = NL West
#204 = NL East
#205 = NL Central
```

```python
mask_leader = all_standings['Games Back'] == '-'
all_standings.loc[mask_leader, 'Games Back'] = 0
all_standings['Games Back'] = all_standings['Games Back'].astype('float64') 
all_standings['Division Rank'] = all_standings['Division Rank'].astype('int64') 

mask200 = all_standings['Division'] == 200
all_standings.loc[mask200,"Division"] = 'AL West'
mask201 = all_standings['Division'] == 201
all_standings.loc[mask201,"Division"] = 'AL East'

mask202 = all_standings['Division'] == 202
all_standings.loc[mask202,"Division"] = 'AL Central'

mask203 = all_standings['Division'] == 203
all_standings.loc[mask203,"Division"] = 'NL West'

mask204 = all_standings['Division'] == 204
all_standings.loc[mask204,"Division"] = 'NL East'

mask205 = all_standings['Division'] == 205
all_standings.loc[mask205,"Division"] = 'NL Central'
```

```python
all_standings['Division'] = all_standings['Division'].astype('category')
```

```python
mask_winners = all_standings['Division Rank'] == 1
mask_2nd_place = all_standings['Division Rank'] == 2
```

```python
division_winners = all_standings[['Season', 'Division', 'Team Name', 'Wins']][mask_winners]
division_2nd_place = all_standings[['Season', 'Division', 'Team Name', 'Wins', 'Games Back']][mask_2nd_place]
```

```python
division_winners.info()
division_2nd_place.info()
```

```python
division_1st_2nd = pd.merge(division_winners, division_2nd_place, on=['Season', 'Division'], how='outer')
```

```python
division_1st_2nd.head(6)
```

```python
def games_to_lose(number):
    return math.floor(number /2)
```

```python
division_1st_2nd['Games to Lose'] = division_1st_2nd['Games Back'].apply(games_to_lose)
```

```python
division_1st_2nd['Needed Wins'] = division_1st_2nd['Wins_x'] - division_1st_2nd['Games to Lose']
```

```python
division_1st_2nd.head(5)
```

```python
labels = division_1st_2nd['Division'].drop_duplicates()
max_wins = []
min_wins = []
mean_wins = []
magic_wins = []
for division in labels:
    division_mask = division_1st_2nd['Division'] == division
    max_wins.append(division_1st_2nd[division_mask]['Wins_x'].max())
    min_wins.append(division_1st_2nd[division_mask]['Wins_x'].min())           
    mean_wins.append(math.floor(division_1st_2nd[division_mask]['Wins_x'].mean()))
    magic_wins.append(division_1st_2nd[division_mask]['Needed Wins'].max())
```

```python
width = 0.35  
fig, ax = plt.subplots()
fig.set_size_inches(15, 10)
max_w = ax.bar(labels, max_wins, label = 'Max Wins')
magic_w = ax.bar(labels, magic_wins, label = 'Magic # Wins')
mean_w = ax.bar(labels, mean_wins, label = 'Mean Wins')
min_w = ax.bar(labels, min_wins, label = 'Minimum Wins')

ax.set_ylim(80,110)
ax.set_ylabel("Wins")
ax.set_title('Wins by Division')
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

for axis in (max_w, magic_w, mean_w, min_w):
    autolabel(axis)
plt.show()
```

```python
x = np.arange(len(labels))  # the label locations
width = .2  # the width of the bars

fig, ax = plt.subplots()
fig.set_size_inches(20, 16)
rects1 = ax.bar(x + width*2, min_wins, width, label='Minimum Wins')
rects2 = ax.bar(x, magic_wins, width, label='Magic # Wins')
rects3 = ax.bar(x + width, mean_wins, width, label='Mean Wins')
rects4 = ax.bar(x - width, max_wins, width, label='Max Wins')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Wins')
ax.set_title('Wins by Division')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()

plt.show()
```

```python
season_labels = division_1st_2nd['Season'].drop_duplicates()
division_labels = division_1st_2nd['Division'].drop_duplicates()

season_mean = []
season_mean_gb = []
for season in season_labels:
    season_mask = division_1st_2nd['Season'] == season
    season_mean.append(round(division_1st_2nd[season_mask]['Wins_x'].mean()))
    season_mean_gb.append(round(division_1st_2nd[season_mask]['Games Back'].mean(),1))
    
mask_al_west= division_1st_2nd['Division'] == 'AL West'
al_west = division_1st_2nd[mask_al_west]['Wins_x']
al_west_gb = division_1st_2nd[mask_al_west]['Games Back']

mask_al_east= division_1st_2nd['Division'] == 'AL East'
al_east = division_1st_2nd[mask_al_east]['Wins_x']
al_east_gb = division_1st_2nd[mask_al_east]['Games Back']

mask_al_central= division_1st_2nd['Division'] == 'AL Central'
al_central = division_1st_2nd[mask_al_central]['Wins_x']
al_central_gb = division_1st_2nd[mask_al_central]['Games Back']

mask_nl_west= division_1st_2nd['Division'] == 'NL West'
nl_west = division_1st_2nd[mask_nl_west]['Wins_x']
nl_west_gb = division_1st_2nd[mask_nl_west]['Games Back']

mask_nl_east= division_1st_2nd['Division'] == 'NL East'
nl_east = division_1st_2nd[mask_nl_east]['Wins_x']
nl_east_gb = division_1st_2nd[mask_nl_east]['Games Back']

mask_nl_central= division_1st_2nd['Division'] == 'NL Central'
nl_central = division_1st_2nd[mask_nl_central]['Wins_x']
nl_central_gb = division_1st_2nd[mask_nl_central]['Games Back']
```

```python
mean_df = pd.DataFrame(season_mean)
mean_df.rename(
    columns = {
        0: 'Mean Wins'},
        inplace =True
)
mean_df['Season'] = seasons
```

```python
division_winners_mean = pd.merge(division_winners, mean_df, on=['Season'], how='outer')
```

```python
def over_under(row):
    if row['Wins'] >= row['Mean Wins']: 
        return 1
    else: return 0 

division_winners_mean['Over Mean'] = division_winners_mean.apply(over_under, axis=1)
```

```python
division_winners_mean.head(5)
```

```python
division_wins_ou_mean=[]

for division in labels:
    division_wins_ou_mean.append(division_winners_mean[division_winners_mean['Division'] == division]['Over Mean'].sum())
```

```python
mean_gb_df = pd.DataFrame(season_mean_gb)
mean_gb_df.rename(
    columns = {
        0: 'Mean Games Back'},
        inplace =True
)
mean_gb_df['Season'] = seasons
```

```python
division_gb_mean = pd.merge(division_2nd_place, mean_gb_df, on=['Season'], how='outer')
```

```python
def over_under_gb(row):
    if row['Games Back'] <= row['Mean Games Back']: 
        return 1
    else: return 0 

division_gb_mean['Under Mean'] = division_gb_mean.apply(over_under_gb, axis=1)
```

```python
division_gb_mean.head(5)
```

```python
division_gb_ou_mean=[]

for division in labels:
    division_gb_ou_mean.append(division_gb_mean[division_gb_mean['Division'] == division]['Under Mean'].sum())
```

# Recap:
## 1) The 'AL West' and 'AL East' are two Division that require the most wins to finish in 1st.
###  - 'Division Winner Wins' were greater than Mean 7 out 10 and 8 out of 10 respectively.
## 2) The 'AL East' and 'NL Central' are two Division that are the most competitive.
### - 'Division Winner Lead' were less than Mean 8 out 10 and 7 out of 10 respectively.
## 3) There is definitely can be a 'Magic # of Wins' for each Division. We saw that each Division had their own unique tendencies
## 4) Looking at World Series Appearances and Winners, even though the 'AL West' and 'AL East' were usually above the 'Wins Mean' in rarely resulting in the ultimate goal of World Series Championship. Once the Playoffs come around, Regular Season record doesn't mean as much
